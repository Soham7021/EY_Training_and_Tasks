```python
import pandas as pd
import matplotlib.pyplot as plt
 
# Sample sales dataset
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Product": ["Laptop", "Laptop", "Laptop", "Mobile", "Mobile", "Mobile"],
    "Units_Sold": [120, 135, 150, 200, 220, 210],
    "Revenue": [600000, 675000, 750000, 300000, 330000, 315000]
}
 
df = pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt

avg_marks = df.groupby("Product")["Revenue"].mean()

avg_marks.plot(kind="bar", color="skyblue", edgecolor="red")

plt.title("avg Revenue by Product")
plt.xlabel("Product")
plt.ylabel("Avg Revenue")
plt.xticks(rotation=0)
plt.show()
```

```python
plt.bar(df["Month"], df["Units_Sold"], color='skyblue')
plt.title("Units Sold per Month")
plt.xlabel("Month")
plt.ylabel("Units Sold")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
for product in df['Product'].unique():
    subset = df[df['Product'] == product]
    plt.bar(subset['Month'], subset['Units_Sold'], label=product)

plt.title('Monthly Units Sold by Product')
plt.xlabel('Month')
plt.ylabel('Units Sold')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
```
