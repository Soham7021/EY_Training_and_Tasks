```python
import pandas as pd
import matplotlib.pyplot as plt
# Sales dataset
data = {
"Month": ["Jan", "Feb", "Mar", "Jan", "Feb", "Mar", "Jan", "Feb", "Mar"],
"Product": ["Mobile", "Mobile", "Mobile", "Laptop", "Laptop", "Laptop", "TV",
"TV", "TV"],
"Units_Sold": [200, 220, 210, 120, 135, 150, 80, 95, 100],
"Revenue": [300000, 330000, 315000, 600000, 675000, 750000, 240000, 285000,
300000]
}
df = pd.DataFrame(data)
df
```

# Step 2: Basic Analysis
## Total revenue per product

```python
total_revenue = df.groupby("Product")["Revenue"].sum()
print("\nTotal Revenue per Product:")
print(total_revenue)
```

# Average units sold per product

```python
average_units_sold = df.groupby("Product")["Units_Sold"].mean()
print("\nAverage Units Sold per Product:")
print(average_units_sold)
```

# Step 3: Visualization 1 – Line Chart

```python
for product in df["Product"].unique():
    monthly_data = df[df["Product"] == product]
    plt.plot(monthly_data["Month"], monthly_data["Revenue"], marker='o', label=product)


plt.title("Monthly Revenue Trend per Product")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Step 4: Visualization 2 – Bar Chart

```python
plt.figure(figsize=(6, 4))
total_revenue.plot(kind='bar', color=['skyblue', 'orange', 'green'])
plt.title("Total Revenue per Product")
plt.xlabel("Product")
plt.ylabel("Total Revenue")
plt.tight_layout()
plt.show()
```

# Step 5: Visualization 3 – Pie Chart

```python
total_revenue.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title("Revenue Contribution per Product")
plt.ylabel("")
plt.tight_layout()
plt.show()
```
